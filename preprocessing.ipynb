```python
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from google.colab import drive
```

```python
# Mount Google Drive
drive.mount('/content/drive')
```

```python
# Example usage:
input_excel_file = '/content/drive/My Drive/FYP/DataPreprocessing/Dataset.xlsx'  # Replace with the path to your Excel file in Google Drive
output_csv_file = '/content/drive/My Drive/FYP/DataPreprocessing/preprocessed_data.csv'  # Define the output CSV file in Google Drive
sequence_length = 5  # Adjust as needed
```

```python
# Step 1: Read Dzongkha text from the Excel file in Google Drive
df = pd.read_excel(input_excel_file)
# Access the first (index 0) column and convert its data to strings
first_column = df.iloc[:, 0].astype(str)

# Join the strings into a single string
dzongkha_text = " ".join(first_column)
```

```python
# Step 2:
# Tokenize the Dzongkha text
tokens = dzongkha_text.split()

# Create input sequences and target words
input_sequences = []
target_words = []
```

```python
for i in range(len(tokens) - sequence_length):
    input_seq = tokens[i:i + sequence_length]
    target_word = tokens[i + sequence_length]
    input_sequences.append(input_seq)
    target_words.append(target_word)
```

```python
# Step 3: Tokenization using Keras Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(input_sequences)
input_sequences = tokenizer.texts_to_sequences(input_sequences)
vocab_size = len(tokenizer.word_index) + 1  # Add 1 for padding token
```

```python
tokens
```

```python
# Step 4: Padding sequences to a fixed length
max_sequence_length = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='pre')
```

```python
# Step 5: Convert target words to numeric IDs
target_word_ids = tokenizer.texts_to_sequences(target_words)
```

```python
target_word
```

```python
# Ensure both lists have the same length
min_length = min(len(input_sequences), len(target_word_ids))
input_sequences = input_sequences[:min_length]
target_word_ids = target_word_ids[:min_length]
```

```python
# Convert input sequences and target word IDs to strings
input_sequences = [" ".join(map(str, seq)) for seq in input_sequences]
target_word_ids = [str(id[0]) if id else '' for id in target_word_ids]  # Handle empty lists
```

```python
# Create a DataFrame with separate columns for input sequences and target word IDs
df_preprocessed = pd.DataFrame({'InputSequences': input_sequences, 'TargetWordIDs': target_word_ids})

# Save the preprocessed data to a CSV file in Google Drive
df_preprocessed.to_csv(output_csv_file, index=False, encoding='utf-8')
print(f'Preprocessed data saved to {output_csv_file} in Google Drive.')
```
